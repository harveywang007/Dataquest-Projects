# Storytelling Data Visualization on Exchange Rates

This is a project from Dataquest. We will visualize some trends in the exchange rates of the euro.

Namely, we will see:
1. how the euro-dollar exchange rate changed during the pandemic
2. how the euro-dollar exchange rate changed during the 2007-2008 financial crisis
3. how the euro-dollar exchange rate changed during the presidencies of Bush, Obama, and Trump.

The dataset we will examine contains exchange rates between 1999 and 2021. This dataset is still being updated as of the creation of this project; our particular dataset only goes up to January 2021. The dataset can be found [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020), but ultimately derives from the European Central Bank.

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

# I will examine a little bit of the dataset
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

There are 5699 rows and 41 columns; one of the columns provides information on the date, whereas the rest are currencies. This means that this dataset contains 40 currencies.

There are quite a few null values, ranging from hundreds per column to thousands per column!

Some columns are strings, whereas others are float. This will mean that we will have to convert the columns, if we are to perform further analysis.

# Data Cleaning

For the purposes of this project, we will only focus on the US dollar and euro.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

# I will print some of the dataset to check
exchange_rates.head()
```

```python
# Now, I will isolate the Time and US_dollar
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

A hyphen appears quite a few times in `US_dollar`, so we must drop it.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# And I will convert the column into a float type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Visualizing the Data

Now that the data is cleaned, we can visualize it with a line plot.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

Of course, there is more to this project! We are not finished at all.

The graph is hardly informative or interesting, at this point. We can see that the line plot is quite "fuzzy" because of the daily variations. However, despite the blurriness, there is a clearer, more general trend in the line plot.

# Visualizing with Rolling Means

We can smooth out the graph by plotting the rolling means of the data.

Below, we will show plots of the data with various rolling windows.

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, window in zip([2, 3, 4, 5, 6],
                     [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(window).mean())
    plt.title('Rolling Window: ' + str(window), weight='bold')
    
plt.tight_layout()
plt.show()
```

We are going to use a rolling window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

# I will take a look at the new dataframe
euro_to_dollar
```

# How the Euro-Dollar Exchange Rate Changed during the Pandemic

In this section, we will see how the coronavirus pandemic affected the exchange rate between the two currencies.

We will use the 2016-2019 data as a baseline. As the pandemic response across the US took a few weeks, for the sake of this project, I will choose March 13, 2020 as the start of the pandemic; this is when President Trump [declared the pandemic to be a national emergency](https://www.ajmc.com/view/covid19-roundup2). Thus, all data from March 13, 2020 onwards will be considered data from the pandemic.

```python
import datetime as dt

recent_years = euro_to_dollar[euro_to_dollar['Time'].dt.year >= 2016]
prepandemic = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2016) &
                             (euro_to_dollar['Time'] < dt.datetime(2020,3,13))]
pandemic = euro_to_dollar[euro_to_dollar['Time'] >= dt.datetime(2020,3,13)]
```

## Finding high and low values

```python
# I will just use this section to print the pandemic values
print('Lowest value during the pandemic: ' + str(pandemic['US_dollar'].min()))
print('Highest value during the pandemic: ' + str(pandemic['US_dollar'].max()))
```

```python
# I will just use this section to print the prepandemic values
print('Lowest value before the pandemic: ' + str(prepandemic['US_dollar'].min()))
print('Highest value before the pandemic: ' + str(prepandemic['US_dollar'].max()))
```

## Visualizing the data

```python
# Using the fivethirtyeight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
plt.figure(figsize=(15,9))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3]

# Adding the y-axis values
for ax in axes:
    ax.set_ylim(1.0, 1.3)
    ax.set_yticks([1.0, 1.1, 1.2, 1.3])
    ax.set_yticklabels(['', '1.1', '1.2', '1.3'], alpha=0.5)

# ax1: prepandemic years
ax1.plot(prepandemic['Time'], prepandemic['rolling_mean'],
        linewidth=1, color='#33A1C9')
ax1.set_xticklabels(['2016', '', '2017', '', '2018', '',
                     '2019', '', '2020'], alpha=0.5) # Setting the tick labels for x-axis
ax1.text(736511, 1.3, s='Prepandemic', weight='bold',
         size=18, color='#33A1C9') # Setting a title
ax1.grid(alpha=0.3)

# ax2: pandemic years
ax2.plot(pandemic['Time'], pandemic['rolling_mean'],
        linewidth=3, color='#af0b1e')
ax2.set_xticklabels(['March 2020', '', '', 'June 2020', '', '',
                     'Sept. 2020', '', '', 'Dec. 2020', ''],
                    alpha=0.5) # Setting the tick labels for x-axis
ax2.text(737607, 1.3, s='Pandemic', weight='bold',
         size=18, color='#af0b1e') # Setting a title
ax2.grid(alpha=0.3)

# ax3: the overall timeline
ax3.plot(recent_years['Time'], recent_years['rolling_mean'],
        linewidth=1, color='#33A1C9')
ax3.plot(pandemic['Time'], pandemic['rolling_mean'],
        linewidth=3, color='#af0b1e') # Highlighting the pandemic
ax3.set_xticklabels(['2016', '2017', '2018', '2019',
                    '2020', '2021'], alpha=0.5) # Setting the tick labels for x-axis
ax3.grid(alpha=0.3)

# Adding a title for the whole visual
ax1.text(735800, 1.38,
        s='Euro-Dollar exchange rate peaked at 1.25 and 1.23 before and during the pandemic',
        weight='bold', size=20)
ax1.text(735800, 1.35,
        s='Euro-Dollar exchange rate prepandemic (2016-2019) and during the pandemic (after March 13, 2020)',
        size=17)

# Adding a signature
ax3.text(735800, 0.94, 'Created by Harvey Wang' + ' '*42 +
         'Special Thanks to Dataquest' + ' '*41 +
         'Source: European Central Bank', color = '#f0f0f0',
         backgroundcolor = '#4d4d4d', size=14)

plt.show()
```

## Conclusion

Surprisingly enough, the highest and lowest exchange rates occurred before the pandemic, in 2017 and 2018, respectively. The exchange rate peaked at 1.25 and dipped at 1.04. The consequent financial crisis created by the pandemic did not increase nor decrease the exchange rate outside of those extremes.

That being said, the exchange rate during the pandemic (up until January 2021) has been quite variable, given how little time has passed since its start. In less than a year, the exchange rate dipped at 1.07 and rose to 1.23. All in all, the prepandemic years had a comparatively more uniform path.

It is important to note that the euro-dollar exchange rate was declining up until the pandemic. It declined for a few days after March 13, 2020 until it started rising. As of January 2021, it was rising.

# How the Euro-Dollar Exchange Rate Changed during the 2007-2008 Financial Crisis

In this section, we will see how the 2007-2008 financial crisis affected the exchange rate between the two currencies.

We will use the 2006 and 2009 data as a comparison.

```python
few_years = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2006) & 
                           (euro_to_dollar['Time'].dt.year < 2010)]
crisis = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2007) & 
                        (euro_to_dollar['Time'].dt.year <= 2008)]
```

## Finding high and low values

```python
# I will just use this section to print the crisis values
print('Lowest value during the crisis: ' + str(crisis['US_dollar'].min()))
print('Highest value during the crisis: ' + str(crisis['US_dollar'].max()))
```

## Visualizing the data

```python
# Using the fivethirtyeight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(few_years['Time'], few_years['rolling_mean'],
        linewidth=1, color='#13831E')

# Highlighting the crisis
ax.plot(crisis['Time'], crisis['rolling_mean'],
        linewidth=3, color='#B11DBF')
ax.axvspan(xmin=733130, xmax=733285, alpha=0.3, color='grey')

# Adding the tick labels
ax.set_xticklabels(['2006', '', '2007', '', '2008', '',
                    '2009', '', '2010'], alpha=0.5) # Setting the tick labels for x-axis
ax.set_yticks([1.15, 1.2, 1.3, 1.4, 1.5, 1.6])
ax.set_yticklabels(['', '1.2', '1.3', '1.4', '1.5', ''], alpha=0.5) # Setting the tick labels for y-axis

# Adding a title
ax.text(732190, 1.63,
        s='Euro-Dollar exchange rate peaked at 1.59 during the 2007-2008 crisis',
        weight='bold', size=20)
ax.text(732190, 1.595,
        s='Euro-Dollar exchange rate between 2006 and 2010',
        size=17)
ax.grid(alpha=0.3)

# Adding a signature
ax.text(732175, 1.08, 'Created by Harvey Wang' + ' '*41 +
         'Special Thanks to Dataquest' + ' '*42 +
         'Source: European Central Bank', color = '#f0f0f0',
         backgroundcolor = '#4d4d4d', size=14)

plt.show()
```

## Conclusion

During the years of 2007 and 2008, the exchange rate increased to a very high value; it reached close to 1.6 twice during those years.

But, even before 2007, there was steady rise in the rate. This rise seemed to continue until the second quarter of 2008. Interestingly enough, after the second peak, the rate dropped rapidly and reached its pre-crisis value at the start of 2009. And, the rate started rising again in 2009.

# How the Euro-Dollar Exchange Rate Changed during the Presidencies of Bush, Obama, and Trump.

In this section, we will see how the presidencies of George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021) has affected the exchange rate.

As presidents are inaugurated on January 20 of the year following their election, I will index my data accordingly. It should be noted that the dataset I am using does not reach January 20 in 2021.

```python
bush = euro_to_dollar[(euro_to_dollar['Time'] >= dt.datetime(2001,1,20)) & 
                      (euro_to_dollar['Time'] < dt.datetime(2009,1,20))]
obama = euro_to_dollar[(euro_to_dollar['Time'] >= dt.datetime(2009,1,20)) & 
                       (euro_to_dollar['Time'] < dt.datetime(2017,1,20))]
trump = euro_to_dollar[euro_to_dollar['Time'] >= dt.datetime(2017,1,20)]
```

## Finding high and low values

```python
# I will just use this section to print the Bush values
print('Lowest value during Bush: ' + str(bush['US_dollar'].min()))
print('Highest value during Bush: ' + str(bush['US_dollar'].max()))
```

```python
# I will just use this section to print the Obama values
print('Lowest value during Obama: ' + str(obama['US_dollar'].min()))
print('Highest value during Obama: ' + str(obama['US_dollar'].max()))
```

```python
# I will just use this section to print the Trump values
print('Lowest value during Trump: ' + str(trump['US_dollar'].min()))
print('Highest value during Trump: ' + str(trump['US_dollar'].max()))
```

## Visualizing the data

```python
# Using the fivethirtyeight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
plt.figure(figsize=(15,9))
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3, ax4]

# Adding the y-axis values
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.5)

# ax1: Bush era
ax1.plot(bush['Time'], bush['rolling_mean'], color='#A2A731')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'], alpha=0.5) # Setting the tick labels for x-axis
ax1.text(731490, 1.57, s='Bush Era', weight='bold', size=18, color='#A2A731') # Setting a title
ax1.text(731440, 1.5, s='(2001-2009)', weight='bold',
         size=15) # Setting a subtitle
ax1.grid(alpha=0.3)

# ax2: Obama era
ax2.plot(obama['Time'], obama['rolling_mean'], color='#306EBF')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'], alpha=0.5) # Setting the tick labels for x-axis
ax2.text(734333, 1.57, s='Obama Era', weight='bold', size=18, color='#306EBF') # Setting a title
ax2.text(734353, 1.5, s='(2009-2017)', weight='bold',
         size=15) # Setting a subtitle
ax2.grid(alpha=0.3)

# ax3: Trump era
ax3.plot(trump['Time'], trump['rolling_mean'], color='#ffa500')
ax3.set_xticklabels(['', '2017', '', '2019', '', '2021'], alpha=0.5) # Setting the tick labels for x-axis
ax3.text(736775, 1.57, s='Trump Era', weight='bold', size=18, color='#ffa500') # Setting a title
ax3.text(736800, 1.5, s='(2009-2017)', weight='bold',
         size=15) # Setting a subtitle
ax3.grid(alpha=0.3)

# ax4: the overall timeline
ax4.plot(bush['Time'], bush['rolling_mean'], color='#A2A731')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#306EBF')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#ffa500')
ax4.set_xticklabels([]) # Removing x-axis labels, as we the other graphs tell the years
ax4.grid(alpha=0.3)

# Adding a title for the whole visual
ax1.text(730120, 1.95,
        s='Euro-Dollar exchange rate peaked at 1.59 and dipped at 0.83',
        weight='bold', size=20)
ax1.text(730120, 1.86,
        s='under the last three presidents',
        weight='bold', size=20)
ax1.text(730120, 1.77,
        s='Euro-Dollar exchange rate under Pres. Bush (2001-2009), Obama (2009-2017), and Trump (2017-2021)',
        size=17)

# Adding a signature
ax4.text(729950, 0.73, 'Created by Harvey Wang' + ' '*45 +
         'Special Thanks to Dataquest' + ' '*45 +
         'Source: European Central Bank', color = '#f0f0f0',
         backgroundcolor = '#4d4d4d', size=14)

plt.show()
```

## Conclusion

Bush's presidency saw the highest and lowest rates during this entire period. Throughout his presidency, there was a steady growth in the rates overall; he also oversaw the financial crisis in 2007 and 2008. However, there was a steep drop near the end of it. The most variation occurred during the Bush era.

Obama's presidency saw a downward trend in the rates overall, but this drop was not as steep as the climb during the Bush era. Obama's presidency also had many rises and falls, along with a steep drop at the end of 2014 and the start of 2015. For the rest of his presidency, the exchange rate remained stable and at a low value.

Trump's presidency saw the least amount of variation. There was a steep rise up until mid-2018, before dropping. The drop was much shallower than the earlier rise. This drop continued until the start of the coronavirus pandemic, where we see another rise, albeit not as high as the highest values under Bush or Obama.
